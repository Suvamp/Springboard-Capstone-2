Distribution of Violent/Nonviolent crimes per state

```python
# Calculate the average Nonviolent & Violent Crimes and sort by state
CrimePerState = crime.groupby(by='State')[['nonViolPerPop', 'ViolentCrimesPerPop']].mean()
CrimePerState.head()
```

Distribution of vacancy per state

```python
VacPerState = crime.groupby(by='statecode')[['HousVacant', 'PctHousOccup', 'PctHousOwnOcc', 'PctVacantBoarded', 'PctVacMore6Mos']].mean()
VacPerState.head()
```

Distribution of Percentage Unemployed/Employed by State

```python
UnempPerState = crime.groupby(by='statecode')[['PctUnemployed', 'PctEmploy']].mean()
UnempPerState.head()
```

Distribution of Education by State

```python
EducPerState = crime.groupby(by='statecode')[['PctLess9thGrade', 'PctNotHSGrad', 'PctBSorMore']].mean()
EducPerState.head()
```

Distribution of Age by State

```python
AgePerState = crime.groupby(by='statecode')[['agePct12t21', 'agePct12t29', 'agePct16t24', 'agePct65up']].mean()
AgePerState.head()
```

```python
# bargraph representing average occurrence of violent & nonviolent crimes per state
(CrimePerState.reindex(index=CrimePerState.mean(axis=1)
    .sort_values(ascending=False)
    .index)
    .plot(kind='barh', figsize=(15, 15), title='Average Violent & Nonviolent Crimes by State'))
plt.xlabel('Per 100K Population');
```

```python
# bargraph representing average percentage of education attained per state
(EducPerState.reindex(index=EducPerState.mean(axis=1)
    .sort_values(ascending=False)
    .index)
    .plot(kind='barh', figsize=(15, 15), title='Average Level of Education Attained by State'))
plt.xlabel('Percentage(%)');
```

```python
# Bargraph representing average percentage of unemployment/employment rate
(UnempPerState.reindex(index=UnempPerState.mean(axis=1)
    .sort_values(ascending=False)
    .index)
    .plot(kind='barh', figsize=(15, 15), title='Average Unemployment/Employment rate by state'))
plt.xlabel('Percentage(%)');
```

```python
crime_occurrence = pd.melt(crime[['State', 'nonViolPerPop', 'ViolentCrimesPerPop']],
                          id_vars = 'State',
                          var_name = 'type of crime',
                          value_vars = ['nonViolPerPop', 'ViolentCrimesPerPop'],
                          value_name = 'number of crimes per 100K population')
```

```python
crime_occurrence.head()
```

```python
plt.subplots(figsize=(15, 15))
sns.boxplot(x='State', y='number of crimes per 100K population', hue='type of crime', data=crime_occurrence)
plt.xticks(rotation='vertical')
plt.ylabel('# of crimes per 100K population')
plt.xlabel('State');
```

```python
vacancy = pd.melt(crime[['State', 'HousVacant']],
                 id_vars = 'State',
                 var_name = 'House Vacancy',
                 value_vars = ['HousVacant'],
                 value_name = 'number of vacant households')
```

```python
vacancy.head()
```

```python
plt.subplots(figsize=(15, 15))
sns.boxplot(x='State', y='number of vacant households', hue='House Vacancy', data=vacancy)
plt.xticks(rotation='vertical')
plt.ylabel('# of vacant households')
plt.xlabel('State');
```
